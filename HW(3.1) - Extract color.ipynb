```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

# Load the image
image = cv2.imread('dataset\colorobject.png')

# Convert the image to RGB color space for Matplotlib
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Define the lower and upper RGB range for yellow color
# Note: These RGB values are a guess; you'll need to adjust them to your specific shade of yellow.
lower_blue = np.array([0, 0, 255])
upper_blue = np.array([0, 255, 255])

# Create a mask to extract yellow objects
blue_mask = cv2.inRange(rgb_image, lower_blue, upper_blue)

# Apply the mask to the original image
blue_objects = cv2.bitwise_and(rgb_image, rgb_image, mask=blue_mask)

# Display the image using Matplotlib
plt.imshow(blue_objects)
plt.show()
```

```python
# Combine the yellow and blue masks to extract both yellow and blue objects
yellow_blue_mask = cv2.bitwise_or(yellow_mask, blue_mask)

# Apply the combined mask to the original image
yellow_blue_objects = cv2.bitwise_and(image, image, mask=yellow_blue_mask)

# Show the result
plt.imshow(cv2.cvtColor(yellow_blue_objects,cv2.COLOR_BGR2RGB))
```
