```python
#set factor
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("dataset/messi.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print(img.shape)

plt.imshow(img)
```

```python
x = 332
y = 280
w = 60
h = 60

res = cv2.getRectSubPix(img,(w,h),(x+(w/2),y+(h/2)))
plt.imshow(res)
```

```python
#set img resolution
res2 = cv2.resize(res,(300,300))
print(res2.shape)
plt.imshow(res2)
```
