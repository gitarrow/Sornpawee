# Example

img tranformation: scale, rotation, translation, shear affine

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
url = 'dataset/flower.jpg'
img = cv2.imread(url)
rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
```

```python
cols, rows, ch = rgb.shape
M = np.float32([[1,0,50], [0,1,50]])

res = cv2.warpAffine(rgb,M,(rows, cols))
print(res.shape)
plt.imshow(res)
```

```python
#Treanslation + schaling
M3 = np.float32([[0.5,0,240], [0,0.5,180]])

res3 = cv2.warpAffine(rgb,M3,(rows, cols))
print(res.shape)
plt.imshow(res3)
```

```python
#Rotation
M4 = cv2.getRotationMatrix2D((0,0),5,1) #(center top left, angle, scale)
print(M4)

res4 = cv2.warpAffine(rgb,M4,(rows, cols)) #(image, M, dsize)
print(res.shape) #Printing the Shape of the Resulting Image
plt.imshow(res4) #Displaying the Rotated Image
```
