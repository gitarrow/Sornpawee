```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('../dataset/coins.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, bw = cv2.threshold(gray, 30, 255, cv2.THRESH_BINARY)

plt.subplot(2,2,1);plt.imshow(gray, cmap='gray')
plt.subplot(2,2,2);plt.imshow(bw, cmap='gray')
print(ret)
```

```python
#dilation
ker = np.ones((3,3),np.uint8)
res = cv2.morphologyEx(bw, cv2.MORPH_DILATE,ker,iterations=3)

plt.imshow(res, cmap = 'gray')
print(ker)
```

```python
dist_euclindian = cv2.distanceTransform(res,cv2.DIST_L2, 5)
dist_cityblock = cv2.distanceTransform(res, cv2.DIST_L2, 5)
dist_chessboard = cv2.distanceTransform(res, cv2.DIST_C, 5)

```

```python
plt.figure(figsize=(18,6))

#Orginal Thrsholded image
plt.subplot(1,4,1)
plt.imshow(res,cmap='gray')
plt.title('Thresholded Image')
plt.axis('off')
#Euclidean Distance transform
plt.subplot(1,4,2)
plt.imshow(dist_euclindian, cmap ='gray')
plt.title('Euclidean Distance')
plt.axis('off')
#City Block Distance transform
plt.subplot(1,4,3)
plt.imshow(dist_cityblock, cmap ='gray')
plt.title('Chessboard Distance')
plt.axis('off')
#Chessboard Distance transform
plt.subplot(1,4,4)
plt.imshow(dist_chessboard, cmap ='gray')
plt.title('Chessboard Distance')
plt.axis('off')
```
