```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
url = 'dataset/dilation.png' #change your image here
img = cv2.imread(url)
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
print(gray.shape)
```

```python
#delate
ksize = 3
iterations =2
ker = np.ones((ksize,ksize),np.uint8)
resErode =cv2.erode(gray, ker, iterations = iterations)
plt.subplot(1,2,1); plt.imshow(gray,cmap='gray')
plt.subplot(1,2,2); plt.imshow(resErode,cmap='gray')
```

```python
result = cv2.merge([resErode,resErode,gray]) #BGR (Blue, Green, Red)
plt.imshow(result)
```

```python
#delate
ksize = 3
iterations =3
ker = np.ones((ksize,ksize),np.uint8)
resErode =cv2.erode(gray, ker, iterations = iterations)
plt.subplot(1,2,1); plt.imshow(gray,cmap='gray')
plt.subplot(1,2,2); plt.imshow(resErode,cmap='gray')
```

```python
letter = cv2.merge([gray,resErode,resErode]) #BGR (Blue, Green, Red)
plt.imshow(letter)
```

```python
# color range in RGB
lowerRGB = np.array([0, 0, 0])  
upperRGB = np.array([0, 0, 0])  

# RGB range
mask = cv2.inRange(letter, lowerRGB, upperRGB)

# Change bg in the mask to green 
green_color = [0, 255, 0]  
letter[mask == 255] = green_color

plt.imshow(letter)
plt.show()
```
