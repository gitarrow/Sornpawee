```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

url = '/content/drive/MyDrive/Year 2 Term 2/Computer Vision/Assignment 1.2/dataset/mario.png'

img = cv2.imread(url)
rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
print(rgb.shape)
```

```python
url_template = '/content/drive/MyDrive/Year 2 Term 2/Computer Vision/Assignment 1.2/dataset/mario_coin.png'

template = cv2.imread(url_template)

rgb_template = cv2.cvtColor(template, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_template)
```

```python
w = template.shape[1]
h = template.shape[0]

print(w,h)

gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
temp = cv2.cvtColor(rgb_template, cv2.COLOR_RGB2GRAY)

res = cv2.matchTemplate(gray,temp,cv2.TM_CCOEFF_NORMED)
print(res.shape)
```

```python
threshold = 0.8
loc = np.where(res >= threshold)

print(*loc[::-1])
print ("5555\n", zip(*loc[::-1]))

for pt in zip(*loc[::-1]):
  print(pt)
  cv2.rectangle(rgb, pt, (pt[0] + w, pt[1] + h), (255,0,0), 2)

  plt.imshow(rgb)
```
