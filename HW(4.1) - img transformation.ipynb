```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image from Google Drive
url = 'dataset/rectangle2.png'
img = cv2.imread(url)

# Convert BGR to RGB
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Select source points (Orgial point) for example, three corners of a rectangle
src_points = np.float32([[50, 50], [200, 50], [50, 200]])

#transformed image(Mapping to New Points)
dst_points = np.float32([[10, 100], [200, 50], [100, 250]])

# Calculate the affine transformation matrix
M = cv2.getAffineTransform(src_points, dst_points)

# Apply the affine transformation
transformed_rgb = cv2.warpAffine(rgb, M, (rgb.shape[1], rgb.shape[0]))

# Display the original and transformed images
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(rgb)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(transformed_rgb)
plt.title('Transformed Image')
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('dataset/right.jpg')

# Orginal points provided
src_pts = np.float32([(358, 110), (530, 140), (315, 335), (450, 415)])

# Target point
dst_pts = np.float32([[0, 0], [300, 0], [0, 420], [300, 420]])
```

```python
M = cv2.getPerspectiveTransform(src_pts, dst_pts)
warped_img = cv2.warpPerspective(img, M, (300, 420)) #The hight the the width  of the image.

warped_img_rgb = cv2.cvtColor(warped_img, cv2.COLOR_BGR2RGB)
plt.imshow(warped_img_rgb)
plt.title('Transformed Image')
plt.show()
```
