```python
url = 'dataset/cmerama 2.png'
img = cv2.imread(url)

#OpenCV base one BGR
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
print()
```

```python
#spllit RGB image to R,G, B ch
R, G, B = cv2.split(rgb)
plt.subplot(2,2,1); plt.imshow(R,cmap='gray')
print(R.shape)

plt.subplot(2,2,2); plt.imshow(G,cmap ='gray')
plt.subplot(2,2,3); plt.imshow(B,cmap ='gray')
```

```python
newing = cv2.merge([R,G,B])
plt.imshow(newing)
```

```python
#convert RGB to gray scale
gray = cv2.cvtColor(rgb,cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap = 'gray')
print(gray.shape)
```

```python
#threshold techique
#gray mode to B&W mode

ret , bw =cv2.threshold(gray,100,255, cv2.THRESH_BINARY_INV)
plt.imshow(bw, cmap = 'gray')
print(ret)
```

```python
#exam q3
#threshold techique
#gray mode to B&W mode

ret , bw =cv2.threshold(gray,100,255, cv2.THRESH_TRUNC)
plt.imshow(bw, cmap = 'gray')
print(ret)
```

```python
bg = bw*0 #This line creates an array bg which is the same size as bw but filled with zeros. It represents a black image.
Green = cv2.merge([bg,bw,bg]) #(r,g,b) merge 3 chanels (not show, show, not show)
plt.imshow(Green)
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image
img = cv2.imread('dataset\colorobject.png')  # Replace with your image path

# Convert the image to RGB
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Define the range for blue color in RGB
# Adjust these values based on the specific blue you want to extract
lower_color = np.array([0, 0, 178])  # Less intense blue in RGB
upper_color = np.array([0, 0, 255])  # More intense blue in RGB

# Create a binary mask where the blue colors are white and everything else is black
mask = cv2.inRange(rgb_img, lower_color, upper_color)

# Apply the mask to the original (RGB) image to extract the blue color
blue_shapes = cv2.bitwise_and(rgb_img, rgb_img, mask=mask)

# Display the result
plt.imshow(blue_shapes)
plt.axis('off')  # Hide the axis
plt.show()
```

```python
lower_color = np.array([0, 100, 0])
upper_color = np.array([80, 255, 80])

# Create a binary mask where the green colors are white and everything else is black
mask = cv2.inRange(img, lower_color, upper_color)

# Apply the mask to the original image to extract the green color
green_shapes = cv2.bitwise_and(img, img, mask=mask)

# Display the result
plt.imshow(cv2.cvtColor(green_shapes, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Hide the axis
plt.show()
```
