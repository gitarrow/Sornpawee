```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
url = 'dataset/coins.jpg' #change your image here
img = cv2.imread(url)
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
# gray =cv2.resize(gray,None,fx=0.1,fy=0.1)
plt.imshow(gray,cmap='gray')
print(gray.shape)
```

```python
#prewitt
kernelX = np.float32(([1,0,-1,],
                      [1,0,-1,],
                      [1,0,-1,],))
kernelY = np.transpose(kernelX)
print(kernelX,'\n\n', kernelY)

Gx = cv2.filter2D(gray,cv2.CV_16S,kernelX)
Gy = cv2.filter2D(gray,cv2.CV_16S,kernelY)

GxA = cv2.convertScaleAbs(Gx)
GyA = cv2.convertScaleAbs(Gy)

plt.subplot (2,2,1);plt.imshow(Gx,cmap='gray')
plt.subplot (2,2,2);plt.imshow(Gy,cmap='gray')
plt.subplot (2,2,3);plt.imshow(GxA,cmap='gray')
plt.subplot (2,2,4);plt.imshow(GyA,cmap='gray')
```

```python
resPrewitt = GxA + GyA
plt.subplot(1,2,1); plt.imshow(gray,cmap='gray')
plt.subplot(1,2,2); plt.imshow(resPrewitt,cmap='gray')
```

```python
#sobel
Gx = cv2.Sobel(gray, cv2.CV_16S,1,0,ksize=3)
Gy = cv2.Sobel(gray, cv2.CV_16S,0,1,ksize=3)

GxA = cv2.convertScaleAbs(Gx)
GyA = cv2.convertScaleAbs(Gy)

plt.subplot(2,2,1); plt.imshow(Gx,cmap='gray')
plt.subplot(2,2,2); plt.imshow(Gy,cmap='gray')
plt.subplot(2,2,3); plt.imshow(GxA,cmap='gray')
plt.subplot(2,2,4); plt.imshow(GyA,cmap='gray')
```

```python
resSobel = GxA+GyA
plt.subplot(1,2,1);plt.imshow(gray,cmap='gray')
plt.subplot(1,2,2);plt.imshow(resSobel,cmap='gray')
```

```python
#laplacian
res = cv2.Laplacian(gray, cv2.CV_32F, ksize = 3)
res1 = cv2.convertScaleAbs(res)

plt.subplot(1,2,1);plt.imshow(gray,cmap='gray')
plt.subplot(1,2,2);plt.imshow(res1,cmap='gray')
```

```python
#Canny
resCanny = cv2.Canny(gray, 50,100)
plt.imshow(resCanny)
```
